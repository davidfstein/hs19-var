```python
import pandas as pd
import numpy as np
import os
import json
```

```python
files_coord_curves = [x for x in os.listdir('./') if x.startswith('coord_curve')]
```

```python
files_coord_curves
```

```python
file_coord_cells = 'coord_cells.csv'
```

```python
num = 50
list_curves = []
for x in files_coord_curves:
    df_file_x = pd.read_csv(x,sep='\t')
    c1 = np.linspace(df_file_x.iloc[0,0], df_file_x.iloc[-1,0], num)
    c2 = np.linspace(df_file_x.iloc[0,1], df_file_x.iloc[-1,1], num)
    c3 = np.linspace(df_file_x.iloc[0,2], df_file_x.iloc[-1,2], num)
    for i in range(num):
        dict_coord_curves = dict()
        dict_coord_curves['branch_id'] = x[-9:-4]
        dict_coord_curves['pseudotime'] = i
        dict_coord_curves['x'] = c1[i]
        dict_coord_curves['y'] = c2[i]
        dict_coord_curves['z'] = c3[i]
        list_curves.append(dict_coord_curves)
```

```python
list_curves
```

```python
with open('curves.json', 'w') as f:
    json.dump(list_curves, f)
```
